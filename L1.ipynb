# Chat API (SDK)

```python
import os
from openai import OpenAI

from dotenv import load_dotenv, find_dotenv
load_dotenv(override=True)

client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL", "https://api.deepseek.com/v1")
)
```

```python
def get_completion(prompt, model="deepseek-chat"):
    messages = [{"role": "user", "content": prompt}]    ## List of dicts as inputs
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, 
    )
    return response.choices[0].message.content  ## Return a complex deeply nested object
```

```python
bag="LangChain"

style="Patient and rigorous"

prompt=f"Explain the use of {bag} and give a short example with {style}"
```

```python
response = get_completion(prompt) 
print(response)
```

## Output Parsers

```python
passage="The LangChain framework is designed to simplify the creation of applications using large language models.\
      It provides standard, extendable interfaces and integrations for various modules, including model I/O, retrieval mechanisms, and agents.\
        By utilizing document loaders, text splitters, and vector embeddings, developers can connect their proprietary or custom data sources to AI models\
        , enabling context-aware and highly accurate responses."

response_template="""For the following text, extract the following information:
is_framework: Is this text introducing a development framework? If yes, answer True; if no or unknown, answer False.
core_modules: What core modules (such as model I/O, agents, etc.) are mentioned in the text? Output them as a comma-separated Python list. If this information is not found, output an empty list [].
data_connections: What components used for connecting custom data sources (such as loaders, splitters, etc.) are mentioned in the text? Output them as a comma-separated Python list. If this information is not found, output an empty list [].

Output in JSON format containing the following three keys:
is_framework
core_modules
data_connections

Text: {text}"""
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_template(response_template)
messages = prompt_template.format_messages(text=passage)

chat = ChatOpenAI(model="deepseek-chat", temperature=0.0,
                  api_key=os.getenv("DEEPSEEK_API_KEY"),
                  base_url=os.getenv("DEEPSEEK_BASE_URL"))
response = chat.invoke(messages)
print(response.content) ## Still str
```

```python
print(type(response.content))
```

```python
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field
from typing import List

class TextExtraction(BaseModel):
    is_framework: bool = Field(description="Is this text introducing a development framework?")
    core_modules: List[str] = Field(description="List of core modules mentioned in the text.")
    data_connections: List[str] = Field(description="List of components used for connecting custom data sources.")

parser = JsonOutputParser(pydantic_object=TextExtraction)
```

```python
chain = prompt_template | chat | parser ## LCEL: prompt|llm|Outputparser

response_dict = chain.invoke({
    "text": passage,
    "format_instructions": parser.get_format_instructions()
})

print(type(response_dict)) 
print(response_dict)
```
